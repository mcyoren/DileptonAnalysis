```python
widthX, widthY, widthZ = [], [], []
for xbin in range(1,1+hists_ee_BG_read[iOption][0].GetNbinsX()):
    widthX.append(hists_ee_BG_read[iOption][0].ProjectionX("widthX").GetBinWidth(xbin))
for ybin in range(1,1+hists_ee_BG_read[iOption][0].GetNbinsY()):
    widthY.append(hists_ee_BG_read[iOption][0].ProjectionY("widthY").GetBinWidth(ybin))
for zbin in range(1,1+hists_ee_BG_read[iOption][0].GetNbinsZ()):
    widthZ.append(hists_ee_BG_read[iOption][0].ProjectionZ("widthZ").GetBinWidth(zbin))

phiwidthX, phiwidthY, phiwidthZ = [], [], []
for xbin in range(1,1+hists_phi_ee_BG_read[iOption][0].GetNbinsX()):
    phiwidthX.append(hists_phi_ee_BG_read[iOption][0].ProjectionX("phiwidthX").GetBinWidth(xbin))
for ybin in range(1,1+hists_phi_ee_BG_read[iOption][0].GetNbinsY()):
    phiwidthY.append(hists_phi_ee_BG_read[iOption][0].ProjectionY("phiwidthY").GetBinWidth(ybin))
for zbin in range(1,1+hists_phi_ee_BG_read[iOption][0].GetNbinsZ()):
    phiwidthZ.append(hists_phi_ee_BG_read[iOption][0].ProjectionZ("phiwidthZ").GetBinWidth(zbin))
```

```python
def GetBinVolume(xbin,ybin,zbin):
    return (widthX[xbin-1]*widthY[ybin-1]*widthZ[zbin-1])**0
def GetBinVolumePhi(xbin,ybin,zbin):
    return (phiwidthX[xbin-1]*phiwidthY[ybin-1]*phiwidthZ[zbin-1])**0
```

```python
m_proj = hists_ee_BG_read[iOption][0].ProjectionX("m_proj")
pt_proj = hists_ee_BG_read[iOption][0].ProjectionZ("pt_proj")

def InIntegrationRange(xbin,ybin,zbin):
    mee = m_proj.GetBinCenter(xbin)
    pt = pt_proj.GetBinCenter(zbin)
    mt = (pt**2 + mee**2)**0.5
    if mee >= 0.3 and mt <= 1.2 and pt - 1.5*mee <= 0.2 and pt - 0.75*mee >= 0:
        return  True
    return False
```
