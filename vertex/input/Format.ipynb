```python
def Format_Hist_total(second, title_x="p_T",  title_y="counts", left=0.15, bottom=0.15, right=0.01, top=0.01,  Tsize=0.07,  Lsize=0.06,\
                      Mstyle=21,  Msize=2, Mcolor=4,  Lwidth=3,  Lcolor=4,  offset_x=1, offset_y=1, title="",  Malpha=1,  Lalpha=1):

    second.SetStats(0)

    pad = root.gPad
    if left>0:pad.SetLeftMargin(left)
    if right>0:pad.SetRightMargin(right)
    if top>0:pad.SetTopMargin(top)
    if bottom>0:pad.SetBottomMargin(bottom)
    pad.SetTickx()
    pad.SetTicky()

    if Mcolor == 3: Mcolor = root.kGreen+2
    if Mcolor == 5: Mcolor = root.kOrange+7
    if Lcolor == 3: Lcolor = root.kGreen+2
    if Lcolor == 5: Lcolor = root.kOrange+7

    second.SetMarkerStyle(Mstyle)
    second.SetMarkerSize(Msize)
    second.SetMarkerColorAlpha(Mcolor,Malpha)
    second.SetTitle(title)
    
    second.SetLineWidth(Lwidth)
    second.SetLineColorAlpha(Lcolor,Lalpha)
    
    xaxis = second.GetXaxis()
    yaxis = second.GetYaxis()
    
    xaxis.SetTitle(title_x)
    xaxis.CenterTitle(1)
    xaxis.SetTitleFont(42)
    xaxis.SetTitleSize(Tsize)
    xaxis.SetTitleOffset(offset_x)
    
    xaxis.SetLabelFont(42)
    xaxis.SetLabelSize(Lsize)
    xaxis.SetTickSize(0.035)
    
    
    yaxis.SetTitle(title_y)
    yaxis.CenterTitle(1)
    yaxis.SetTitleFont(42)
    yaxis.SetTitleSize(Tsize)
    yaxis.SetTitleOffset(offset_y)
    
    yaxis.SetLabelFont(42)
    yaxis.SetLabelSize(Lsize)
    yaxis.SetTickSize(0.035)
    
    #second.SetTitle(title)
    #second.SetTitleFont(42)
    second.SetFillColorAlpha(0,0)

    #second.Draw()
    
    
```

```python
def Format_Graph( g,  Mstyle=21,  Msize=1,  Mcolor=1,  Lwidth=1,  Lcolor=1,  Malpha=1,  Lalpha=1):
    g.SetMarkerStyle(Mstyle)
    g.SetMarkerSize(Msize)
    g.SetMarkerColorAlpha(Mcolor,Malpha)
    g.SetTitle("")
    #if Mcolor == 3: g.SetMarkerColorAlpha(kGreen + 2,Malpha)
    #if Mcolor == 5: g.SetMarkerColorAlpha(kOrange + 7,Malpha)
    g.SetLineWidth(Lwidth)
    g.SetLineColorAlpha(Lcolor,Lalpha)
    #if Lcolor == 3: g.SetLineColorAlpha(kGreen + 2,Lalpha)
    #if Lcolor == 5: g.SetLineColorAlpha(kOrange + 7,Lalpha)
```

```python
def Format_Graph_Shade( g, x, low, up, Lcolor,  Lalpha):
    g.SetTitle("")
    N = len(low)
    for i in range(N):
        if low[i] < up[i]:
            low[i] = 0
            up[i] = 0
            
    for i in range(N):
            g.SetPoint(i,x[i],low[i]);
            g.SetPoint(N + i, x[N-i-1],up[N-i-1]);
        
    g.SetFillStyle(1001);
    g.SetFillColorAlpha(Lcolor, Lalpha);
    
    #g.SetLineWidth(Lwidth)
    #g.SetLineColorAlpha(Lcolor,Lalpha)
    #if Lcolor == 3: g.SetFillColorAlpha(kGreen + 2,Lalpha)
    #if Lcolor == 5: g.SetFillColorAlpha(kOrange + 7,Lalpha)
```

```python
def DrawTypeC( syst,  abc,  width,  Lwidth,  Lcolor,  Lalpha):
    
    xxx = array( 'd', [abc + 0.0, 1e30])
    
    yyy = array( 'd', [1, 1e30])
    
    del_yyy =  array( 'd', [syst + 0.0, 0])
    
    del_xxx =  array( 'd', [width + 0.0 , 0])
    
    gr_typeC = root.TGraphErrors(1,xxx,yyy,del_xxx,del_yyy);
    gr_typeC.SetMarkerSize(0)
    if (Lalpha == 0):
        gr_typeC.SetLineWidth(Lwidth)
        gr_typeC.SetLineColor(Lcolor)
        gr_typeC.SetFillStyle(0)
    else:    
        gr_typeC.SetLineWidth(0)
        gr_typeC.SetFillColorAlpha(Lcolor,Lalpha)
    
    #gr_typeC.Draw("2P")
    
    return gr_typeC
   
```

```python
def Format_Pad_old( left,  right,  min_boarder,  max_boarder,  title_x,  title_y,  offset_x,  offset_y,  Tsize,  Lsize, title, lefty=0.15, bottom=0.15, righty=0, top=0, name = "sec"):
    second = root.TH1F(name, name, 100, left, right)
    second.SetMinimum(min_boarder)
    second.SetMaximum(max_boarder)
    second.SetStats(0)
    
    pad = root.gPad
    if lefty>0:pad.SetLeftMargin(lefty)
    if righty>0:pad.SetRightMargin(righty)
    if top>0:pad.SetTopMargin(top)
    if bottom>0:pad.SetBottomMargin(bottom)
    pad.SetTickx()
    pad.SetTicky()
    
    xaxis = second.GetXaxis()
    yaxis = second.GetYaxis()
    
    xaxis.SetTitle(title_x)
    xaxis.CenterTitle(1)
    xaxis.SetTitleFont(42)
    xaxis.SetTitleSize(Tsize)
    xaxis.SetTitleOffset(offset_x)
    
    xaxis.SetLabelFont(42)
    xaxis.SetLabelSize(Lsize)
    xaxis.SetTickSize(0.035)
    
    
    yaxis.SetTitle(title_y)
    yaxis.CenterTitle(1)
    yaxis.SetTitleFont(42)
    yaxis.SetTitleSize(Tsize)
    yaxis.SetTitleOffset(offset_y)
    
    yaxis.SetLabelFont(42)
    yaxis.SetLabelSize(Lsize)
    yaxis.SetTickSize(0.035)
    
    second.SetTitle(title)
    second.SetTitleFont(42)
    second.SetFillColorAlpha(0,0)

    second.Draw()
    
    return second
```

```python
def Unity_Line(left, right, scale =1):
    levelwide = root.TLine()
    levelwide.SetLineWidth(1)
    levelwide.SetLineColor(1)
    levelwide.SetLineStyle(7)
    
    levelwide.DrawLine(left, scale, right, scale)
```

```python
def Format_Func( f,  LineWidth, LineColor,  LineStyle=1,  Lalpha=1):
    f.SetLineColorAlpha(LineColor,Lalpha)
    f.SetLineWidth(LineWidth)
    f.SetLineStyle(LineStyle)
```

```python
def Format_Legend( legend, TextSize=0.05, align =22, LineWidth=0, Color=1,  alpha=0):
    legend.SetTextAlign(align)
    legend.SetTextSize(TextSize)
    legend.SetFillColorAlpha(Color,alpha)
    legend.SetLineWidth(LineWidth)
    #legend.SetLineColorAlpha(Color,alpha)
```

```python
def Draw_Legend_Title(x1,y1,x2,y2,title,TextSize=0.05, align =22, LineWidth=0, Color=1,  alpha=0):
    legend = root.TLegend(x1,y1,x2,y2,title)
    legend.SetHeader(title)
    legend.SetTextAlign(align)
    legend.SetTextSize(TextSize)
    legend.SetFillColorAlpha(Color,alpha)
    legend.SetLineWidth(LineWidth)
    return legend
    #legend.SetLineColorAlpha(Color,alpha)
```
