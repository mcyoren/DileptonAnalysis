```python
%run input/Format.ipynb
import ROOT as root
from array import array
root.gErrorIgnoreLevel = root.kFatal
%jsroot on
```

```python
iOption0 = 1
sim = 1
part = 0
N_files = 2
```

```python
N_centr = 5
N_pt = 1
colors=[1,2,4,root.kGreen+2,root.kMagenta,root.kOrange+4,root.kGray,root.kCyan]
central_bins = [0,100,0,20,20,40,40,60,60,80,80,100]
pt_boarders = [0.5,0.9,0.2,0.5,0.5,1.0,1.0,5.0,0.5,5]
file_path="input/rg/"
file_names=["my-27th_QA_19200_350runs"]
hist_select_3D_names = ["ep_hist","n0_hist","prob_hist","disp_hist","chi2npe0_hist"]
hist_e_select_3D_names = ["ep_hist_el","n0_hist_el","prob_hist_el","disp_hist_el","chi2npe0_hist_el","rich_prob1","rich_prob2","rich_prob3"]
hist_ghost_select_3D_names=["el_had_dphi","el_had_dz","el_had_dr"]
hist_pt = ["el_pt_hist"]
hists_all = [hist_select_3D_names,hist_e_select_3D_names,hist_ghost_select_3D_names,hist_pt]
```

```python
if sim :
    file_path="../sim/output/embed/"
    file_names=["electron_2M_hagedorn_v3_QA","piminus_2M_hagedorn_v3_QA","piplus_2M_hagedorn_v3_QA"]

    #phi_bins = [96,97,98,99,100,100,101,101,102,103,104,105]
N_type = len(hists_all)
```

```python
hists_read = []

for iFile in range(N_files):
    print(file_path+file_names[iFile]+".root")
    infile = root.TFile.Open(file_path+file_names[iFile]+f".root", "read")
    hists_emc_read0 = []
    for itype in range(N_type):
        hist_select_3D0 = []
        for ihist in range(len(hists_all[itype])):
            hist_select_3D0.append(infile.Get(hists_all[itype][ihist]))
            hist_select_3D0[-1].SetDirectory(root.nullptr)
            hist_select_3D0[-1].SetName(hist_select_3D0[-1].GetName()+f"{iFile}")
        hists_emc_read0.append(hist_select_3D0)
    hists_read.append(hists_emc_read0)

infile.Close()
```

```python
icentr = 0
type_loc = 0 
N_type_loc= len(hists_all[type_loc])
c0 = root.TCanvas(f"c0",f"c0",1400,450*N_type_loc)
c0.Divide(2,N_type_loc)
for itype in range(N_type_loc):
    for iFile in range(0,N_files):
        c0.cd(itype*N_files+iFile+1)
        ibin0 = hists_read[iFile][type_loc][itype].ProjectionZ().FindBin(central_bins[2*icentr+0])
        ibin1 = hists_read[iFile][type_loc][itype].ProjectionZ().FindBin(central_bins[2*icentr+1])
        hists_read[iFile][type_loc][itype].GetZaxis().SetRange(ibin0,ibin1)
        hists_read[iFile][type_loc][itype].Project3D("yx").Draw("colz")

c0.Draw()
c0.SaveAs(f"output/rich_emcal_qa/qa_pt_centr_{icentr}.png")
```

```python
icentr = 2
ipt = 0
type_loc = 3
N_type_loc= len(hists_all[type_loc])
c0 = root.TCanvas(f"c4",f"c4",1400,900)
c0.Divide(2,2)

ibinX0 = hists_read[iFile][type_loc][itype].ProjectionX().FindBin(pt_boarders[2*ipt+0])
ibinX1 = hists_read[iFile][type_loc][itype].ProjectionX().FindBin(pt_boarders[2*ipt+1])

ibinZ0 = hists_read[iFile][type_loc][itype].ProjectionZ().FindBin(central_bins[2*icentr+0])
ibinZ1 = hists_read[iFile][type_loc][itype].ProjectionZ().FindBin(central_bins[2*icentr+1])

hist_e = hists_read[0][type_loc][0].ProjectionY("hist_e",ibinX0,ibinX1,ibinZ0,ibinZ1)
hist_d = hists_read[0][type_loc][0].ProjectionY("hist_d",ibinX0,ibinX1,ibinZ0,ibinZ1)
hist_r = hists_read[0][type_loc][0].ProjectionY("hist_r",ibinX0,ibinX1,ibinZ0,ibinZ1)
hist_h = hists_read[1][type_loc][0].ProjectionY("hist_h",ibinX0,ibinX1,ibinZ0,ibinZ1)


hist_e.Scale(1./(hist_e.GetBinContent(1)+1))
hist_d.Scale(1./(hist_d.GetBinContent(1)+1))
hist_r.Scale(1./(hist_r.GetBinContent(1)+1))
hist_h.Scale(1./(hist_h.GetBinContent(1)+1))

c0.cd(1)
hist_e.GetXaxis().SetRange(6,15)
hist_e.Draw("H")

c0.cd(2)
hist_d.Divide(hist_h)
hist_d.GetXaxis().SetRange(6,15)
hist_d.Draw()

c0.cd(3)
hist_h.GetXaxis().SetRange(6,15)
hist_h.Draw()

c0.cd(4)
hist_r.Divide(hist_h)
hist_r.Divide(hist_h)
hist_r.GetXaxis().SetRange(6,15)
hist_r.Draw()

c0.Draw()
c0.SaveAs(f"output/rich_emcal_qa/pt_ratio_{icentr}_{ipt}.png")
```
