```python
N_centr = 6
N_options = 6
N_pt = 1
colors=[1,2,4,root.kGreen+2,root.kMagenta,root.kOrange+4,root.kGray,root.kCyan]
central_bins = [0,20,60,80,20,40,40,60,60,80,80,100]
pt_boarders = [0.4,0.7,1.0,2.0,0,5.0,1.0,5.0,0.5,5]
file_path="../ML/output/"
file_names=["BDT_test.root","BDT_test1.root"]
N_Files = len(file_names)
part_names = ["b#bar{b}","c#bar{c}","J/#psi","#phi","#gamma #rightarrow ee","#pi^{-}","#pi^{-}","#gamma #rightarrow ee"]
hist_names = ["e_pt_hist_AI", "h_pt_hist_AI", "e_pt_hist_SC", "h_pt_hist_SC", "e_pt_hist_NC", "h_pt_hist_NC"]
N_type = len(hist_names)
```

```python
hists_read = []

for iFile in range(N_Files):
    infile = root.TFile.Open(file_path+file_names[iFile], "read")
    hists_emc_read0 = []
    for itype in range(N_type):
        hists_emc_read0.append(infile.Get(hist_names[itype]))
        hists_emc_read0[-1].SetDirectory(root.nullptr)
        hists_emc_read0[-1].SetName(hists_emc_read0[-1].GetName()+f"{iFile}")
        hists_emc_read0[-1].Scale(1e-6)
    hists_read.append(hists_emc_read0)

infile.Close()
```
