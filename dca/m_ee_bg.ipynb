```python
new_DCA_bins = array( 'd', [0,25,50,75,100,125,150,200,250,300,350,400,500,1000])
#new_DCA_bins = array( 'd', [0,25,50,75,100,125,150,175,225,275,325,375,425,525,600,1000])
def rebin_to_edges(h, edges, make_density=True):
    """Variable-bin rebin. Returns a NEW hist. Optionally converts to dN/dx."""
    original_bin_widths = h.GetBinWidth(1)  # assuming uniform binning in original hist
    h2 = h.Rebin(len(edges)-1, h.GetName()+"_reb", edges)  # <-- NEW hist
    h2.Sumw2()  # keep proper errors if you ever scale/add later
    
    if make_density:
        for i in range(1, h2.GetNbinsX()+1):
            w = h2.GetBinWidth(i)
            if w > 0:
                h2.SetBinContent(i, h2.GetBinContent(i)/w*original_bin_widths)
                h2.SetBinError(i,   h2.GetBinError(i)/w*original_bin_widths)
    return h2
```

```python
c1 = root.TCanvas(f"c1",f"c1",900,900)
c1.Divide(1,1)
legends =[]
if_divide = False
alphas = [0.5,0.3,0.3,0.3,0.3,0.3,0.3,0.3,0.3,0.3,0.3,0.3,0.3]

legend = root.TLegend(0.55,0.5,0.99,0.99); legends.append(legend)
for ifile in range(len(file_names)):
    c1.cd(1)
    proj = hists_read[ifile][2]
    proj_syst = hists_read[ifile][3]

    proj      = rebin_to_edges(proj,      new_DCA_bins, make_density=False)
    proj_syst = rebin_to_edges(proj_syst, new_DCA_bins, make_density=False)
    proj1 = proj.Rebin(1,proj.GetName()+"_reb")
    proj_syst1 = proj_syst.Rebin(1,proj_syst.GetName()+"_reb")
    proj1.Smooth(1)
    proj_syst1.Smooth(1)
    proj.Add(proj1,0.99)
    proj_syst.Add(proj_syst1,0.99)
    proj.Scale(2./4)
    proj_syst.Scale(2./4)
    
    proj.GetXaxis().SetRangeUser(0,499)
    #proj.Getyaxis().SetRangeUser(1,499)
    
    Format_Hist_total(proj,"DCA_{T}^{e^{+}e^{-}} [#mum]","dN/DCA_{T}^{e^{+}e^{-}} [#mum^{-1}]",0.17,0.13,0.01,0.01,0.07,0.06,21,1.5,4,3,4,0.85,1.24)
    proj.SetMarkerColorAlpha(colors0[ifile],(10-1)/10.); proj.SetLineColorAlpha  (colors0[ifile],(10-1)/10.)
    proj.SetMinimum(3e-11);proj.SetMaximum(9.9e-8)
    root.gPad.SetLogy(); 
    if ifile == 0:  proj.Draw("")   
    legends[-1].SetBorderSize(0); legends[-1].SetFillStyle(0)
    legends[-1].AddEntry(proj,Opt_names[ifile],"lep")
    legends[-1].Draw()
    for ic in range(ifile,len(file_names)):
        c1.cd(1+ic)
        proj_syst.SetFillColorAlpha(colors0[ifile],alphas[ifile-ic])
        if ifile == 0 and ic !=0: proj.Draw("")
        else: proj.Draw("same")
        proj_syst.Draw("same e2")
c1.Draw()
c1.SaveAs("output/bg/dca_bg.pdf")
```
