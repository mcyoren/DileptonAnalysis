```python
def copycat_hist(hist,seed=0):
    seed += root.gRandom.Integer(10000)
    new_hist = root.TH3D(hist.GetName()+f"{seed}", hist.GetTitle()+f"{seed}", hist.GetNbinsX(), hist.GetXaxis().GetBinLowEdge(1), hist.GetXaxis().GetBinUpEdge(hist.GetNbinsX()),\
        hist.GetNbinsY(), hist.GetYaxis().GetBinLowEdge(1), hist.GetYaxis().GetBinUpEdge(hist.GetNbinsY()),\
        hist.GetNbinsZ(), hist.GetZaxis().GetBinLowEdge(1), hist.GetZaxis().GetBinUpEdge(hist.GetNbinsZ()))
    for ix in range(1, hist.GetNbinsX()+1):
        for iy in range(1, hist.GetNbinsY()+1):
            for iz in range(1, hist.GetNbinsZ()+1):
                new_hist.SetBinContent(ix, iy, iz, hist.GetBinContent(ix, iy, iz))
    return new_hist
```

```python
import ROOT as root

def make_data_dca_hist(m_lo, m_hi,
                       icentr=0,
                       iOpt=0,
                       rebin=1,
                       pt_low=1, pt_high=12,
                       keff_bg=0.03,
                       sub_koeff=sub_koeff,
                       nev_bin=2):
    """
    Build DCA 1D hist for data in a given mass window [m_lo, m_hi],
    using the same logic as your DCA macro:

      - project FG and BG from hists_read_3d
      - rebin
      - convert to per-m (divide by bin width)
      - scale by 1/N_events (hists_read_3d[iOpt][2].GetBinContent(nev_bin))
      - subtract (sub_koeff - keff_bg) * BG, i.e. FG - 0.97 * BG for sub_koeff=1, keff_bg=0.03

    Returns: TH1D with FG - 0.97*BG normalized per event and per m.
    """

    # 3D histograms: [0]=FG, [1]=BG, [2]=N_events vs centrality
    h3_fg = hists_read_3d[iOpt][0]
    h3_bg = hists_read_3d[iOpt][1]

    # mass axis is X
    ax_m = h3_fg.GetXaxis()
    bin_low = ax_m.FindBin(m_lo)
    bin_up  = ax_m.FindBin(m_hi)

    # protect edges
    if m_lo <= ax_m.GetBinLowEdge(1):
        bin_low = 1
    if m_hi >= ax_m.GetBinUpEdge(ax_m.GetNbins()):
        bin_up = ax_m.GetNbins()

    # project DCA (Y axis), integrate pT (Z axis) over [pt_low, pt_high]
    name_fg = f"dca_stat_{icentr}_{m_lo:.3f}_{m_hi:.3f}"
    name_bg = f"dca_BG_{icentr}_{m_lo:.3f}_{m_hi:.3f}"

    proj    = h3_fg.ProjectionY(name_fg, bin_low, bin_up, pt_low, pt_high)
    proj_bg = h3_bg.ProjectionY(name_bg, bin_low, bin_up, pt_low, pt_high)

    proj.Rebin(rebin)
    proj_bg.Rebin(rebin)

    # per-bin Poisson errors + divide by bin width (per m)
    for ibin in range(1, proj.GetNbinsX()+1):
        bw = proj.GetBinWidth(ibin)

        c_fg = proj.GetBinContent(ibin)
        e_fg = (abs(c_fg))**0.5

        c_bg = proj_bg.GetBinContent(ibin)
        e_bg = (abs(c_bg))**0.5

        # convert to per m
        if bw > 0:
            proj.SetBinContent(ibin, c_fg / bw)
            proj.SetBinError(ibin,   e_fg / bw)

            proj_bg.SetBinContent(ibin, c_bg / bw)
            proj_bg.SetBinError(ibin,   e_bg / bw)

    # 1/N_events normalization (same as your code)
    nev = hists_read_3d[iOpt][2].GetBinContent(nev_bin)
    if nev > 0:
        proj.Scale(1.0/nev)
        proj_bg.Scale(sub_koeff/nev)  # this is the BG used for subtraction

    # subtract (1 - keff_bg)*BG   for sub_koeff=1 this is FG - 0.97*BG
    proj.Add(proj_bg, -1.0 + keff_bg)

    # optional: style in your way here if you want
    # Format_Hist_total(...)

    proj.SetDirectory(0)
    return proj
```

```python
def make_sim_dca_templates(m_lo, m_hi,
                           icentr=0,
                           sim_dca=None,
                           shift=0,
                           rebin=1,
                           pt_low=1, pt_high=25,
                           keffs=None,
                           skip_thermal=False):
    """
    Build DCA templates for simulation components in sim_dca
    for a given mass window [m_lo, m_hi], using the same logic as your macro:

      - project DCA (X) in [m_lo, m_hi] on Y axis (mass axis in sim is Y)
      - sum 5 subsamples (jsim=1..5) with scale factor 10/DCA * keffs[isim]
      - normalize by total h_stat entries
      - rebin
      - (optionally skip thermal)

    Returns: dict {part_names[isim]: TH1D}
    """

    if sim_dca is None:
        # default: use your global list
        sim_dca = [1, 3, 4, 5, 8]

    if keffs is None:
        keffs = [1,1,1,1,1e-7,0.25e-7,1,1,1.5e-7,1,1,1]

    templates = {}

    # we use one of the components just to get the mass axis
    ref_isim = sim_dca[0]
    h3_ref = hists_sim[ref_isim][icentr]
    ax_m_sim = h3_ref.GetYaxis()  # in sim: Y = mass

    sim_bin_low = ax_m_sim.FindBin(m_lo)
    sim_bin_up  = ax_m_sim.FindBin(m_hi)

    for isim in sim_dca:
        if skip_thermal and isim == 8:
            continue

        name_base = hists_sim[isim][icentr].GetName()
        name_proj = f"{name_base}_proj_dca_{m_lo:.3f}_{m_hi:.3f}"

        # base projection for this component
        sim_proj = hists_sim[isim][icentr].ProjectionX(
            name_proj,
            sim_bin_low, sim_bin_up,
            pt_low, pt_high
        )
        sim_proj.SetDirectory(0)
        sim_proj.Scale(0.0)

        # stats hist used to normalize
        h_stat = copycat_hist(hists_sim[isim][icentr+18])
        h_stat.Scale(0.0)

        # sum 5 subsamples as in your macro
        for jsim in range(1, 6):
            h3_sub = hists_sim[isim][shift+jsim]
            name_sub = f"{h3_sub.GetName()}_proj_dca_{isim}_{jsim}"
            sim_sub = h3_sub.ProjectionX(
                name_sub,
                sim_bin_low, sim_bin_up,
                pt_low, pt_high
            )

            # scale factor from your code: 10/DCA * keffs[isim]
            bw = sim_proj.GetBinWidth(1)
            if bw <= 0:
                bw = 1.0

            scale = 10.0 / bw * keffs[isim]
            sim_proj.Add(sim_sub, scale)

            h_stat.Add(hists_sim[isim][jsim+18])

        # normalize by number of events
        if h_stat.GetEntries() > 0:
            sim_proj.Scale(1.0 / h_stat.GetEntries())

        if rebin > 1:
            sim_proj.Rebin(rebin)

        # you may also want per-m like data:
        # for ibin in range(1, sim_proj.GetNbinsX()+1):
        #     bw = sim_proj.GetBinWidth(ibin)
        #     if bw > 0:
        #         sim_proj.SetBinContent(ibin, sim_proj.GetBinContent(ibin)/bw)

        templates[part_names[isim]] = sim_proj

    return templates
```

```python
import numpy as np

def fit_window_thermal_only(m_lo, m_hi, m_center,
                            icentr=0,
                            isim_thermal=8,
                            sim_dca=None,
                            shift=0,
                            rebin=1,
                            lambda_prior=2.0,
                            k_min=0.1, k_max=300.0):
    """
    Fit ONLY the thermal component in this mass window [m_lo, m_hi].

    Data:    FG - 0.97 * BG          (make_data_dca_hist)
    Sim:     DCA templates           (make_sim_dca_templates)
    Free:    one parameter k_th multiplying the thermal template.
    Others:  fixed to their nominal keffs.

    Returns: k_th (float).
    """

    if sim_dca is None:
        sim_dca = [1, 3, 4, 5, 8]   # phi, ccbar, bbar, thermal, etc.

    # -----------------------------
    # 1) Data histogram (signal)
    # -----------------------------
    h_data = make_data_dca_hist(m_lo, m_hi,
                                icentr=icentr,
                                iOpt=0,
                                rebin=rebin,
                                keff_bg=0.03,   # 3% under-subtraction
                                sub_koeff=sub_koeff_or)
    h_data.SetDirectory(0)

    nbins = h_data.GetNbinsX()
    y   = np.array([h_data.GetBinContent(i) for i in range(1, nbins+1)], dtype=float)
    err = np.array([h_data.GetBinError(i)   for i in range(1, nbins+1)], dtype=float)

    w = np.zeros_like(err)
    mask_err = err > 0
    w[mask_err] = 1.0 / (err[mask_err]**2)
    if mask_err.sum() < 3:
        return 1.0

    # -----------------------------
    # 2) Sim templates
    # -----------------------------
    templates = make_sim_dca_templates(m_lo, m_hi,
                                       icentr=icentr,
                                       sim_dca=sim_dca,
                                       shift=shift,
                                       rebin=rebin,
                                       keffs=keffs,
                                       skip_thermal=False)

    therm_name = part_names[isim_thermal]
    if therm_name not in templates:
        return 1.0

    h_th = templates[therm_name]
    T_th = np.array([h_th.GetBinContent(i) for i in range(1, nbins+1)], dtype=float)

    # sum of all other components (fixed)
    F_fixed = np.zeros_like(T_th)
    for isim in sim_dca:
        if isim == isim_thermal:
            continue
        name = part_names[isim]
        if name not in templates:
            continue
        h_c = templates[name]
        F_fixed += np.array([h_c.GetBinContent(i) for i in range(1, nbins+1)], dtype=float)

    # effective data after subtracting fixed components
    y_eff = y - F_fixed

    mask = mask_err
    if mask.sum() < 3:
        return 1.0

    y_use = y_eff[mask]
    T_use = T_th[mask]
    w_use = w[mask]

    if np.all(T_use == 0):
        return 1.0

    # -----------------------------
    # 3) Analytic solution with prior k_th ~ 1
    # -----------------------------
    a = np.sum(w_use * T_use * T_use)
    b = np.sum(w_use * T_use * y_use)

    if a + lambda_prior == 0:
        return 1.0

    k_th = (b + lambda_prior * 1.0) / (a + lambda_prior)

    # enforce non-negative and clip to [k_min, k_max]
    k_th = float(np.clip(k_th, k_min, k_max))
    return k_th
```

```python
import numpy as np

# mass_centers_thermal: np.array of window centers
# k_th_raw: np.array of raw thermal k_eff per window

def smooth_keff_thermal(mass_centers, k_raw,
                        lambda_smooth=5.0,
                        k_min=0.2, k_max=150):
    """
    Smooth k_raw(m) by minimizing:
      sum_i (k_i - k_raw_i)^2 + lambda * sum_i (k_{i+1} - k_i)^2
    and keep average close to 1.
    """
    m = np.array(mass_centers, dtype=float)
    k = np.array(k_raw,        dtype=float)
    N = len(k)
    if N < 3:
        return k.copy()

    A = np.zeros((N, N), dtype=float)
    b = k.copy()

    # data term
    for i in range(N):
        A[i, i] += 1.0

    # smoothness term
    for i in range(N-1):
        A[i,   i]   += lambda_smooth
        A[i+1, i+1] += lambda_smooth
        A[i,   i+1] -= lambda_smooth
        A[i+1, i]   -= lambda_smooth

    k_s = np.linalg.solve(A, b)

    # keep global average  1
    avg_raw = np.mean(k)
    avg_sm  = np.mean(k_s)
    if avg_sm > 0:
        k_s *= (avg_raw / avg_sm)

    # clip to reasonable band
    k_s = np.clip(k_s, k_min, k_max)
    return k_s

# smooth your raw thermal k_eff(m)
keff_thermal_smooth = smooth_keff_thermal(m_centers, k_th_raw,
                                          lambda_smooth=5.0,
                                          k_min=0.2, k_max=150)

def keff_thermal_at_mass(m_val):
    """
    Get k_eff^thermal at arbitrary mass m_val via interpolation
    over the smoothed thermal curve.
    """
    if len(m_centers) == 0:
        return 1.0

    m_arr = m_centers
    k_arr = keff_thermal_smooth

    m_clamped = float(np.clip(m_val, m_arr.min(), m_arr.max()))
    return float(np.interp(m_clamped, m_arr, k_arr))
```

```python
def allowed_components(m_center):
    """Decide which components to fit at this mass center."""
    allowed = []
    for isim in sim_dca:
        name = part_names[isim]
        if name.lower().startswith('phi'):
            if phi_lo <= m_center <= phi_hi:
                allowed.append(name)
        elif name.lower().startswith('jpsi') or 'j/#psi' in name.lower():
            if jpsi_lo <= m_center <= jpsi_hi:
                allowed.append(name)
        else:
            # ccbar, bbar, thermal always allowed
            allowed.append(name)
    return allowed


def fit_window_root(start_bin, end_bin, m_center):
    """
    Do TFractionFitter template fit for a single mass window.
    Returns dict {component_name: (keff, keff_err)}.
    """
    data_hist = make_data_dca_hist(start_bin, end_bin, sub_koeff=sub_koeff_or)
    templates_all = make_sim_dca_templates(start_bin, end_bin)

    allowed = allowed_components(m_center)
    if len(allowed) == 0:
        return {}

    # Build TObjArray of templates
    arr = root.TObjArray()
    active_names = []
    for name in allowed:
        h = templates_all[name]
        # clone & ensure positive contents
        h_cl = h.Clone(f"{h.GetName()}_clone")
        for ibin in range(1, h_cl.GetNbinsX()+1):
            if h_cl.GetBinContent(ibin) < 0:
                h_cl.SetBinContent(ibin, 0.)
        arr.Add(h_cl)
        active_names.append(name)

    # Clone data too
    data_cl = data_hist.Clone(f"{data_hist.GetName()}_clone")

    fitter = root.TFractionFitter(data_cl, arr, "Q")  # "Q" = quiet

    # Optionally: constrain fractions close to 1 (not strictly needed)
    # We expect keff ~ 1; but TFractionFitter returns fractions
    # that sum to 1 by default if not rescaled. Since our templates are normalized per event,
    # the fitted fractions are effectively keffs up to a common normalization.
    # You can leave unconstrained or use something like:
    # for ipar in range(len(active_names)):
    #     fitter.Constrain(ipar, 0., 5.)

    status = fitter.Fit()
    if status != 0:
        print(f"Fit failed for window m={m_center:.3f}, status {status}")
        return {}

    results = {}
    for ipar, name in enumerate(active_names):
        val = root.Double(0.)
        err = root.Double(0.)
        fitter.GetResult(ipar, val, err)
        # These "val" are the template fractions in the mixture.
        # If you have already normalized templates to your expected integrals
        # (via keffs[]), then 'val' is effectively your keff(m).
        results[name] = (float(val), float(err))

    return results
```

```python
from sklearn.linear_model import Ridge
import numpy as np

def fit_window_sklearn(start_bin, end_bin, m_center,
                       alpha=0.1,
                       k_min=0.1, k_max=15):
    """
    Weighted least-squares template fit using sklearn (Ridge).
    Returns: dict {component_name: keff(m_center)}.
    If the fit is impossible, returns keff=1 for all allowed components.
    """
    data_hist = make_data_dca_hist(start_bin, end_bin, sub_koeff=sub_koeff_or)
    templates_all = make_sim_templates(start_bin, end_bin)

    allowed = allowed_components(m_center)    # make sure this includes ccbar, bbar, thermal everywhere
    if not allowed:
        return {}

    nb = data_hist.GetNbinsX()

    # data vector and errors
    y   = np.array([data_hist.GetBinContent(i) for i in range(1, nb+1)], dtype=float)
    err = np.array([data_hist.GetBinError(i)   for i in range(1, nb+1)], dtype=float)

    # weights: w = 1/, but 0 where =0
    w = np.zeros_like(err)
    mask_err = err > 0
    w[mask_err] = 1.0 / (err[mask_err]**2)

    # design matrix X: columns = template bin contents
    cols  = []
    names = []
    for name in allowed:
        h = templates_all[name]
        col = np.array([h.GetBinContent(i) for i in range(1, nb+1)], dtype=float)
        cols.append(col)
        names.append(name)

    X = np.stack(cols, axis=1)  # shape: (nbins, ncomp)

    # mask out bins where ALL templates are zero OR weight is zero
    mask_templates = np.any(X > 0, axis=1)
    mask = mask_err & mask_templates

    # if too few bins survive, return keff=1 for all allowed components
    if mask.sum() < max(3, len(names)):   # at least 3 points
        return {name: 1.0 for name in names}

    X_use = X[mask]
    y_use = y[mask]
    w_use = w[mask]

    # apply weights via sqrt(w)
    sqrtw = np.sqrt(w_use)
    Xw = X_use * sqrtw[:, None]
    yw = y_use * sqrtw

    # Ridge with non-negative coeffs so templates can't go negative
    model = Ridge(alpha=alpha, fit_intercept=False, positive=True)
    model.fit(Xw, yw)
    coeffs = model.coef_  # one per template

    # clip coefficients so sim cannot differ by >50%
    coeffs = np.clip(coeffs, k_min, k_max)

    return {names[i]: float(coeffs[i]) for i in range(len(names))}
```

```python
def smooth_keff_component(mass_centers, k_raw, w_raw,
                          lambda_smooth=5.0,
                          k_min=0.5, k_max=1.5):
    """
    Smooth k_raw(m) using:
      sum w (k - k_raw)^2 + lambda * sum (k_{i+1}-k_i)^2
    Then rescale so the weighted average stays 1.
    mass_centers, k_raw, w_raw: 1D arrays (same length).
    """
    m = np.array(mass_centers, dtype=float)
    k = np.array(k_raw,       dtype=float)
    w = np.array(w_raw,       dtype=float)

    N = len(k)
    # if everything is zero weight, just return 1 everywhere
    if np.all(w == 0):
        return np.ones_like(k)

    # build A and b of size N x N
    A = np.zeros((N, N), dtype=float)
    b = np.zeros(N,         dtype=float)

    # data term: w_i (k_i - k_i_raw)^2
    for i in range(N):
        A[i, i] += w[i]
        b[i]    += w[i] * k[i]

    # smoothness term: lambda (k_{i+1}-k_i)^2
    for i in range(N - 1):
        A[i,   i]   += lambda_smooth
        A[i+1, i+1] += lambda_smooth
        A[i,   i+1] -= lambda_smooth
        A[i+1, i]   -= lambda_smooth

    k_smooth = np.linalg.solve(A, b)

    # preserve weighted average ~1 (only using bins with any information)
    mask_info = (w > 0)
    num = np.sum(w[mask_info])
    den = np.sum(w[mask_info] * k_smooth[mask_info])
    if den > 0:
        scale = num / den
        k_smooth *= scale

    # clip to [k_min, k_max]
    k_smooth = np.clip(k_smooth, k_min, k_max)

    return k_smooth
```

```python
def smooth_keff_thermal(mass_centers, k_raw,
                        lambda_smooth=5.0,
                        k_min=0.5, k_max=1.5):
    """
    Smooth thermal keff(m) using:
      sum (k_i - k_raw_i)^2 + lambda * sum (k_{i+1}-k_i)^2
    Then rescale so average stays ~1 and clip to [k_min, k_max].
    """
    m = np.array(mass_centers, dtype=float)
    k = np.array(k_raw,        dtype=float)

    N = len(k)
    if N < 3:
        return k.copy()

    # build A and b
    A = np.zeros((N, N), dtype=float)
    b = k.copy()  # since weight=1 for all points now

    # data term
    for i in range(N):
        A[i, i] += 1.0

    # smoothness term
    for i in range(N-1):
        A[i,   i]   += lambda_smooth
        A[i+1, i+1] += lambda_smooth
        A[i,   i+1] -= lambda_smooth
        A[i+1, i]   -= lambda_smooth

    k_s = np.linalg.solve(A, b)

    # keep average ~1
    avg_raw = np.mean(k)
    avg_sm  = np.mean(k_s)
    if avg_sm > 0:
        scale = avg_raw / avg_sm
        k_s *= scale

    # clip
    k_s = np.clip(k_s, k_min, k_max)

    return k_s
```

```python
import numpy as np

def fit_window_thermal_only(start_bin, end_bin, m_center,
                            isim_thermal=8,
                            lambda_prior=10.0,
                            k_min=0.1, k_max=15.0):
    """
    Fit ONLY the thermal component in this mass window with a prior k~1.
    Other components are kept at keff=1 and subtracted from data.

    Returns: keff_thermal(m_center) as float.
    """

    data_hist = make_data_dca_hist(start_bin, end_bin)
    templates_all = make_sim_templates(start_bin, end_bin)

    nb = data_hist.GetNbinsX()

    # data vector and errors
    y   = np.array([data_hist.GetBinContent(i) for i in range(1, nb+1)], dtype=float)
    err = np.array([data_hist.GetBinError(i)   for i in range(1, nb+1)], dtype=float)

    # weights w = 1/sigma^2
    w = np.zeros_like(err)
    mask_err = err > 0
    w[mask_err] = 1.0 / (err[mask_err]**2)

    therm_name = part_names[isim_thermal]
    if therm_name not in templates_all:
        return 1.0

    # thermal template
    T_th = np.array([templates_all[therm_name].GetBinContent(i)
                     for i in range(1, nb+1)], dtype=float)

    # sum of non-thermal templates (k=1)
    T_fixed_sum = np.zeros(nb, dtype=float)
    for name, h in templates_all.items():
        if name == therm_name:
            continue
        T_fixed_sum += np.array([h.GetBinContent(i) for i in range(1, nb+1)], dtype=float)

    # effective data after subtracting fixed components
    y_eff = y - T_fixed_sum

    # use all bins with non-zero error
    mask = mask_err
    if mask.sum() < 3:
        return 1.0

    w_use = w[mask]
    y_use = y_eff[mask]
    T_use = T_th[mask]

    # compute sums for analytic solution
    a = np.sum(w_use * T_use * T_use)      #  w T^2
    b = np.sum(w_use * T_use * y_use)      #  w T y

    # if template has no support, return 1
    if a == 0 and lambda_prior == 0:
        return 1.0

    # solution with prior k~1
    k = (b + lambda_prior * 1.0) / (a + lambda_prior)

    # clip to allowed range
    k = float(np.clip(k, k_min, k_max))

    return k
```

```python
import numpy as np

def thermal_keff_from_integrals(mass_windows,
                                isim_thermal=8,
                                k_min=0.1, k_max=15.0):
    """
    For each mass window (start_bin..end_bin), compute
        k_th(m) = [I_data - sum_{non-th} I_template] / I_th_template
    where integrals are over DCA and pT in that window.
    Returns: mass_centers, k_raw_thermal as numpy arrays.
    """
    therm_name = part_names[isim_thermal]

    k_raw = []
    m_centers = []

    for (start_bin, end_bin, m_lo, m_hi, m_center) in mass_windows:
        # data DCA histogram in this mass window (same as before)
        h_data = make_data_dca_hist(start_bin, end_bin)
        I_data = h_data.Integral()

        # all templates in this window with their nominal scalings
        temps = make_sim_templates(start_bin, end_bin)

        # thermal template integral
        if therm_name not in temps:
            # if missing, we can't correct it  treat as keff=1
            k_raw.append(1.0)
            m_centers.append(m_center)
            continue

        I_th = temps[therm_name].Integral()

        # sum of all other components (fixed at keff=1)
        I_fixed = 0.0
        for name, h in temps.items():
            if name == therm_name:
                continue
            I_fixed += h.Integral()

        # desired thermal yield from data
        I_th_needed = I_data - I_fixed

        # if template has ~no yield, just keep keff=1
        if I_th <= 0 or I_th_needed <= 0:
            k = 1.0
        else:
            k = I_th_needed / I_th

        # clip to reasonable band
        k = float(np.clip(k, k_min, k_max))

        k_raw.append(k)
        m_centers.append(m_center)

    return np.array(m_centers, dtype=float), np.array(k_raw, dtype=float)
```

```python
def smooth_keff_thermal(mass_centers, k_raw,
                        lambda_smooth=5.0,
                        k_min=0.1, k_max=15.0):
    """
    Smooth k_raw(m) by minimizing:
      sum (k_i - k_raw_i)^2 + lambda * sum (k_{i+1}-k_i)^2
    and keep average close to 1.
    """
    m = np.array(mass_centers, dtype=float)
    k = np.array(k_raw,        dtype=float)
    N = len(k)
    if N < 3:
        return k.copy()

    A = np.zeros((N, N), dtype=float)
    b = k.copy()

    # data term
    for i in range(N):
        A[i,i] += 1.0

    # smoothness term
    for i in range(N-1):
        A[i,  i]   += lambda_smooth
        A[i+1,i+1] += lambda_smooth
        A[i,  i+1] -= lambda_smooth
        A[i+1,i]   -= lambda_smooth

    k_s = np.linalg.solve(A, b)

    # keep global average close to 1
    avg_raw = np.mean(k)
    avg_sm  = np.mean(k_s)
    if avg_sm > 0:
        k_s *= (avg_raw / avg_sm)

    # clip
    k_s = np.clip(k_s, k_min, k_max)
    return k_s

keff_smooth_thermal = smooth_keff_thermal(mass_centers, keff_raw_thermal,
                                          lambda_smooth=5.0,
                                          k_min=0.1, k_max=15.0)
```

```python
import numpy as np

def keff_at_mass_isim(isim, m_val):
    """
    Get k_eff for component 'isim' at mass m_val using the smoothed
    k_eff(m) curve. If we don't have a curve for that source, return 1.
    """
    name = part_names[isim]
    if name not in keff_smooth:
        return 1.0

    m_arr = np.array(mass_centers, dtype=float)
    k_arr = np.array(keff_smooth[name], dtype=float)

    if len(m_arr) == 0:
        return 1.0

    # clamp to range of fitted masses
    m_val_clamped = np.clip(m_val, m_arr.min(), m_arr.max())
    return float(np.interp(m_val_clamped, m_arr, k_arr))
```

```python
from sklearn.linear_model import Ridge
import numpy as np

def fit_window_sklearn(start_bin, end_bin, m_center, alpha=0.1):
    """
    Weighted least-squares template fit using sklearn (Ridge).
    Returns: dict {component_name: keff(m_center)}.
    """
    data_hist = make_data_dca_hist(start_bin, end_bin)
    templates_all = make_sim_templates(start_bin, end_bin)

    allowed = allowed_components(m_center)
    if not allowed:
        return {}

    nb = data_hist.GetNbinsX()

    # data vector and errors
    y   = np.array([data_hist.GetBinContent(i) for i in range(1, nb+1)], dtype=float)
    err = np.array([data_hist.GetBinError(i)   for i in range(1, nb+1)], dtype=float)

    # weights: w = 1/, but 0 where =0
    w = np.zeros_like(err)
    mask_err = err > 0
    w[mask_err] = 1.0 / (err[mask_err]**2)

    # design matrix X: columns = template bin contents
    cols  = []
    names = []
    for name in allowed:
        h = templates_all[name]
        col = np.array([h.GetBinContent(i) for i in range(1, nb+1)], dtype=float)
        cols.append(col)
        names.append(name)

    X = np.stack(cols, axis=1)  # shape: (nbins, ncomp)

    # mask out bins where ALL templates are zero OR weight is zero
    mask_templates = np.any(X > 0, axis=1)
    mask = mask_err & mask_templates

    # if too few bins survive, skip this window
    if mask.sum() < len(names):
        # print(f"Skipping window at m={m_center:.3f}: not enough bins ({mask.sum()})")
        return {}

    X_use = X[mask]
    y_use = y[mask]
    w_use = w[mask]

    # apply weights via sqrt(w)
    sqrtw = np.sqrt(w_use)
    Xw = X_use * sqrtw[:, None]
    yw = y_use * sqrtw

    # Ridge with non-negative coeffs so templates can't go negative
    model = Ridge(alpha=alpha, fit_intercept=False, positive=True)
    model.fit(Xw, yw)
    coeffs = model.coef_  # one per template

    return {names[i]: float(coeffs[i]) for i in range(len(names))}
```

```python
keff_raw = {part_names[isim]: [] for isim in sim_dca}
mass_centers = []

for (start_bin, end_bin, m_lo, m_hi, m_center) in mass_windows:
    res = fit_window_sklearn(start_bin, end_bin, m_center, alpha=0.1)
    if not res:
        continue  # skip windows where fit was impossible

    mass_centers.append(m_center)

    for isim in sim_dca:
        name = part_names[isim]
        if name in res:
            keff_raw[name].append(res[name])
        else:
            # window where this source is not allowed (phi / jpsi out of range)
            keff_raw[name].append(0.0)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

mass_centers_arr = np.array(mass_centers).reshape(-1,1)

def smooth_keff(name, degree=1):
    """
    Smooth keff(m) for component 'name' by fitting delta_k = keff - 1
    with a polynomial of given degree in m.
    """
    y = np.array(keff_raw[name], dtype=float)

    # Treat zero entries as "no information" (forbidden windows)
    mask = y > 0
    if mask.sum() < degree + 1:
        return mass_centers_arr.flatten(), y  # not enough points to fit

    X = mass_centers_arr[mask]
    y_valid = y[mask]

    poly = PolynomialFeatures(degree=degree, include_bias=True)
    X_poly = poly.fit_transform(X)

    # Fit delta_k = k_eff - 1, to keep them near 1
    y_delta = y_valid - 1.0
    reg = LinearRegression()
    reg.fit(X_poly, y_delta)

    X_poly_all = poly.transform(mass_centers_arr)
    y_delta_pred = reg.predict(X_poly_all)
    y_smooth = 1.0 + y_delta_pred

    return mass_centers_arr.flatten(), y_smooth


keff_smooth = {}
for isim in sim_dca:
    name = part_names[isim]
    x, y_s = smooth_keff(name, degree=1)  # start with linear; degree=2 if needed
    keff_smooth[name] = (x, y_s)
```
