```python
import numpy as np

def make_pt_edges():
    edges = []
    # 0.3 to 1.0 in 0.1
    edges += list(np.arange(0.3, 1.0 + 1e-9, 0.1))
    # 1.0 to 2.0 in 0.2 (avoid duplicating 1.0)
    edges += list(np.arange(1.2, 2.0 + 1e-9, 0.2))
    # 2.0 to 5.0 in 0.5 (avoid duplicating 2.0)
    edges += list(np.arange(2.5, 5.0 + 1e-9, 0.5))
    return np.array(edges, dtype=np.float32)

pt_edges = make_pt_edges()
npt = len(pt_edges)-1
print("pt_edges:", pt_edges)
print("Npt bins =", len(pt_edges)-1)
```

```python
import numpy as np

def th1_to_edges_cdf_trim_nsigma(h, nsigma=5.0, clamp_negative=True, min_keep_bins=20):
    ax = h.GetXaxis()
    nb = ax.GetNbins()

    # raw arrays
    edges = np.array([ax.GetBinLowEdge(1+i) for i in range(nb)] + [ax.GetBinUpEdge(nb)], dtype=np.float64)
    w = np.array([h.GetBinContent(1+i) for i in range(nb)], dtype=np.float64)
    if clamp_negative:
        w[w < 0] = 0.0

    # mean/rms (ROOT uses weights)
    mu = float(h.GetMean())
    rms = float(h.GetRMS())
    if rms <= 0:
        # fallback: no shape
        cdf = np.zeros(nb+1, dtype=np.float32); cdf[-1] = 1.0
        return edges.astype(np.float32), cdf

    lo_x = mu - nsigma*rms
    hi_x = mu + nsigma*rms

    # bin indices to keep (0-based bins)
    centers = 0.5*(edges[:-1] + edges[1:])
    keep = (centers >= lo_x) & (centers <= hi_x)

    # ensure at least some bins kept
    if keep.sum() < min_keep_bins:
        # widen window if needed
        keep[:] = True

    w_trim = w.copy()
    w_trim[~keep] = 0.0

    s = w_trim.sum()
    if s <= 0:
        cdf = np.zeros(nb+1, dtype=np.float32); cdf[-1] = 1.0
        return edges.astype(np.float32), cdf

    w_trim /= s
    cdf = np.zeros(nb+1, dtype=np.float64)
    cdf[1:] = np.cumsum(w_trim)
    cdf[-1] = 1.0

    return edges.astype(np.float32), cdf.astype(np.float32)
```

```python
import ROOT as root
import numpy as np

def sample_from_edges_cdf(edges, cdf, rng):
    u = rng.Uniform()
    lo, hi = 0, len(cdf)-1
    while hi - lo > 1:
        mid = (lo + hi)//2
        if cdf[mid] >= u: hi = mid
        else: lo = mid
    c0, c1 = float(cdf[lo]), float(cdf[hi])
    x0, x1 = float(edges[lo]), float(edges[hi])
    t = 0.0 if c1 <= c0 else (u - c0)/(c1 - c0)
    return x0 + t*(x1 - x0)

def make_mc_hist_like(href, edges, cdf, n=200000, name="h_mc", seed=12345):
    ax = href.GetXaxis()
    nb = ax.GetNbins()
    hmc = root.TH1D(name, "MC from CDF;DCA;counts", nb, ax.GetXmin(), ax.GetXmax())
    rng = root.TRandom3(seed)
    for _ in range(n):
        x = sample_from_edges_cdf(edges, cdf, rng)
        hmc.Fill(x)
    if hmc.Integral() > 0:
        hmc.Scale(href.Integral()/hmc.Integral())
    return hmc

# -------------------------
# choose what to test
# -------------------------
icent_test = 12
ipt_test   = 7      # choose pT bin index
charge     = +1     # +1 pos, -1 neg

# same z-range you used for projections
zlow, zhigh = 21, 25

# compute Y-axis bin range corresponding to this pt bin
proj_y = hists_read[0][10].GetYaxis()
pt_lo = float(pt_edges[ipt_test])
pt_hi = float(pt_edges[ipt_test+1])
by_lo = proj_y.FindBin(pt_lo + 1e-6)
by_hi = proj_y.FindBin(pt_hi - 1e-6)

# pick correct source hist (pos vs neg)
if charge > 0:
    h3 = hists_read[0][icent_test + 0]
else:
    h3 = hists_read[0][icent_test + 5]

# rebuild reference projection with the SAME cuts as used to build tables
hist_ref = h3.ProjectionX("href", by_lo, by_hi, zlow, zhigh)

# fetch your sampled-table
edges, cdf = tables[(icent_test, ipt_test, charge)]
hmc = make_mc_hist_like(hist_ref, edges, cdf, n=200000, name="hmc")

# draw
c = root.TCanvas("c_check", "CDF sampling check", 1200, 900)
root.gPad.SetLogy()
hist_ref.SetLineColor(1)
hmc.SetLineColor(2)

hist_ref.SetTitle("DCA sanity check;DCA;counts")
hist_ref.Draw("HIST")
hmc.Draw("HIST SAME")

leg = root.TLegend(0.55, 0.75, 0.88, 0.88)
leg.AddEntry(hist_ref, "Original ProjectionX", "l")
leg.AddEntry(hmc, "MC sampled from trimmed CDF", "l")
leg.Draw()

c.Draw()

print("Tested: icent =", icent_test,
      "ipt =", ipt_test,
      f"pt=[{pt_lo:.2f},{pt_hi:.2f}] GeV",
      "charge =", ("pos" if charge>0 else "neg"))
```

```python
def print_cxx_header_from_tables_pt(tables, cent_list, pt_edges,
                                   header_name="DcaTables_cent_pt"):
    """
    tables key: (icent, ipt, charge) -> (dca_edges, cdf)
      charge: +1 (pos), -1 (neg)

    Exports:
      pt_edges[NPT+1]
      dca_edges[NB+1]
      cdf_pos[NCENT][NPT][NB+1]
      cdf_neg[NCENT][NPT][NB+1]
    """

    ncent = len(cent_list)
    npt   = len(pt_edges) - 1

    # pick reference binning from first available key
    edges_ref = tables[(cent_list[0], 0, +1)][0]
    nb = len(edges_ref) - 1

    # --- header ---
    print(f"#ifndef {header_name.upper()}_H")
    print(f"#define {header_name.upper()}_H\n")
    print(f"struct {header_name} {{")
    print(f"  enum {{ NCENT = {ncent}, NPT = {npt}, NB = {nb} }};")
    print(f"  static const float pt_edges[NPT+1];")
    print(f"  static const float dca_edges[NB+1];")
    print(f"  static const float cdf_pos[NCENT][NPT][NB+1];")
    print(f"  static const float cdf_neg[NCENT][NPT][NB+1];")
    print("};\n")

    # --- pt edges ---
    print(f"const float {header_name}::pt_edges[{npt+1}] = {{")
    print("  " + ", ".join([f"{float(x):.7g}f" for x in pt_edges]) + "};\n")

    # --- dca edges ---
    print(f"const float {header_name}::dca_edges[{nb+1}] = {{")
    print("  " + ", ".join([f"{float(x):.7g}f" for x in edges_ref]) + "};\n")

    # --- cdf blocks ---
    def print_cdf(which):
        print(f"const float {header_name}::cdf_{which}[{ncent}][{npt}][{nb+1}] = {{")
        for ic, icentr in enumerate(cent_list):
            print("  {")
            for ipt in range(npt):
                cdf = tables[(icentr, ipt, +1 if which=="pos" else -1)][1].copy()
                cdf[-1] = 1.0  # enforce exact 1

                row = ", ".join([f"{float(v):.7g}f" for v in cdf])
                comma = "," if ipt != npt-1 else ""
                print(f"    {{ {row} }}{comma}")
            comma_c = "," if ic != ncent-1 else ""
            print(f"  }}{comma_c}")
        print("};\n")

    print_cdf("pos")
    print_cdf("neg")

    print(f"#endif // {header_name.upper()}_H")
```

```python
import numpy as np
import ROOT as root

def cdf_to_quantiles(edges, cdf, nq=256):
    """
    edges: float array length NB+1
    cdf:   float array length NB+1, monotonic, cdf[0]=0, cdf[-1]=1
    returns qs: float array length nq
    """
    qs = np.zeros(nq, dtype=np.float32)
    targets = (np.arange(nq, dtype=np.float64) + 0.5) / float(nq)

    # ensure last exactly 1
    cdf = np.array(cdf, dtype=np.float64, copy=True)
    cdf[0]  = 0.0
    cdf[-1] = 1.0

    for j, u in enumerate(targets):
        lo, hi = 0, len(cdf)-1
        while hi - lo > 1:
            mid = (lo + hi)//2
            if cdf[mid] >= u: hi = mid
            else: lo = mid

        c0, c1 = float(cdf[lo]), float(cdf[hi])
        x0, x1 = float(edges[lo]), float(edges[hi])
        t = 0.0 if c1 <= c0 else (u - c0) / (c1 - c0)
        qs[j] = x0 + t*(x1 - x0)

    return qs
```

```python
def sample_from_quantiles(qs, rng):
    nq = len(qs)
    u = rng.Uniform() * nq
    i = int(u)
    if i <= 0:      return float(qs[0])
    if i >= nq-1:   return float(qs[nq-1])
    t = u - i
    return float(qs[i]*(1.0-t) + qs[i+1]*t)

def make_mc_hist_like_quantiles(href, qs, n=200000, name="h_mcq", seed=12345):
    ax = href.GetXaxis()
    nb = ax.GetNbins()
    hmc = root.TH1D(name, "MC from quantiles;DCA;counts", nb, ax.GetXmin(), ax.GetXmax())
    rng = root.TRandom3(seed)
    for _ in range(n):
        hmc.Fill(sample_from_quantiles(qs, rng))
    if hmc.Integral() > 0:
        hmc.Scale(href.Integral()/hmc.Integral())
    return hmc

icent_test = 12

hpos_ref = hists_read[0][icent_test+0].ProjectionX("hpos_ref", binlow, binhigh, zlow, zhigh)
hneg_ref = hists_read[0][icent_test+5].ProjectionX("hneg_ref", binlow, binhigh, zlow, zhigh)

hpos_mc = make_mc_hist_like_quantiles(hpos_ref, qtables_1to2[(icent_test, +1)], n=300000, name="hpos_mc", seed=1)
hneg_mc = make_mc_hist_like_quantiles(hneg_ref, qtables_1to2[(icent_test, -1)], n=300000, name="hneg_mc", seed=2)

c = root.TCanvas("c_qcheck_charge","Quantile sampling check (pos/neg)",1200,900)
root.gPad.SetLogy()

hpos_ref.SetLineColor(1); hpos_mc.SetLineColor(2)
hneg_ref.SetLineColor(6); hneg_mc.SetLineColor(4)

hpos_ref.Draw("HIST")
hpos_mc.Draw("HIST SAME")
hneg_ref.Draw("HIST SAME")
hneg_mc.Draw("HIST SAME")

leg = root.TLegend(0.55, 0.72, 0.88, 0.88)
leg.AddEntry(hpos_ref, "pos original", "l")
leg.AddEntry(hpos_mc,  "pos MC(quant)", "l")
leg.AddEntry(hneg_ref, "neg original", "l")
leg.AddEntry(hneg_mc,  "neg MC(quant)", "l")
leg.Draw()

c.Draw()
```

```python
def print_cxx_header_quantiles(
    qtables, cent_list, pt_edges, nq=256,
    header_name="DcaQTables_cent10to14",
    fallback_value=0.0
):
    ncent = len(cent_list)
    npt   = len(pt_edges) - 1

    def get_q(icentr, ipt, charge):
        key = (icentr, ipt, charge)
        if key in qtables:
            qs = qtables[key]
            if qs is not None and len(qs) == nq:
                return qs
        # fallback: all quantiles = fallback_value
        return np.full(nq, float(fallback_value), dtype=np.float32)

    # -------- header guard + struct --------
    #print(f"#ifndef {header_name.upper()}_H")
    #print(f"#define {header_name.upper()}_H\n")
    print(f"struct {header_name} {{")
    print(f"  enum {{ NCENT = {ncent}, NPT = {npt}, NQ = {nq} }};")
    print(f"  static const int   cent_values[NCENT];")
    print(f"  static const float pt_edges[NPT+1];")
    print(f"  static const float q_pos[NCENT][NPT][NQ];")
    print(f"  static const float q_neg[NCENT][NPT][NQ];")
    print("};\n")

    # -------- cent values --------
    print(f"const int {header_name}::cent_values[{ncent}] = {{")
    print("  " + ", ".join([str(int(c)) for c in cent_list]) + "};\n")

    # -------- pt edges --------
    print(f"const float {header_name}::pt_edges[{npt+1}] = {{")
    print("  " + ", ".join([f"{float(x):.3g}" for x in pt_edges]) + "};\n")

    # -------- dump tables --------
    def dump(which):
        print(f"const float {header_name}::q_{which}[{ncent}][{npt}][{nq}] = {{")
        for ic, icentr in enumerate(cent_list):
            print("  {")
            for ipt in range(npt):
                qs = get_q(icentr, ipt, +1 if which=="pos" else -1)
                row = ", ".join([f"{float(v):.3g}" for v in qs])
                comma = "," if ipt != npt-1 else ""
                print(f"    {{ {row} }}{comma}")
            comma_c = "," if ic != ncent-1 else ""
            print(f"  }}{comma_c}")
        print("};\n")

    dump("pos")
    dump("neg")

    #print(f"#endif // {header_name.upper()}_H")
```
